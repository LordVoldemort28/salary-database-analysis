```
#imports
import pandas as pd
import os
import gender_guesser.detector as gender
import re
```

```
!ls ./sample_data/
```

```
salaryRoaster = pd.read_excel('./SalaryDataAnalysis/data/2019-2020-Salaries.xlsx')
salaryRoaster
```

```
names = salaryRoaster['Employee']
names
```

```
d = gender.Detector(case_sensitive=False)
genderList = []
for name in names:
  first_name = name.split(", ")[1].split(" ")[0]
  # print(first_name)
  genderList.append(d.get_gender(first_name))
  # print(first_name+ ": " + d.get_gender(first_name))

genderList = ["male" if (x == "mostly_male") else x for x in genderList]
genderList = ["female" if (x == "mostly_female") else x for x in genderList]

salaryRoaster['Gender'] = genderList
print(len(genderList))
print(salaryRoaster.groupby('Gender').nunique())
print(salaryRoaster[salaryRoaster.Gender == 'unknown'].shape[0])
```

```
salaryRoaster.to_csv("./SalaryDataAnalysis/data/salaryWithGender.csv")
```
